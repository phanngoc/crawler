```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://www.nhatot.com/mua-ban-dat-da-nang?page=2"

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

print(soup.prettify())

# Find all listings on the page
listings = soup.find_all("div", class_="AdBody_adItem__1z9l2")

# Extract relevant information for each listing
for listing in listings:
    title = listing.find("h3", class_="commonStyle_adTitle__g520j").text.strip()
    area = listing.find("div", class_="AdBody_adUpperBody__4Mt0b").text.strip()
    price = listing.find("div", class_="AdBody_adItemPrice__Xr5sP").text.strip()

    print(f"Title: {title}")
    print(f"Area: {area}")
    print(f"Price: {price}")
    print("-" * 30)
```

```python
import csv
from selenium import webdriver
from bs4 import BeautifulSoup

DRIVER_PATH = '/Users/ngocp/.pyenv/versions/3.10.3/lib/python3.10/site-packages/selenium/webdriver'

# def create_driver():
#     options = Options()
#     options.add_argument("--window-size=1920,1200")
#     options.add_argument('--no-sandbox')
#     options.add_argument('--headless')

#     driver = webdriver.Chrome(options=options)
#     return driver

# Initialize webdriver
driver = webdriver.Chrome(DRIVER_PATH)  # replace with the path to your chromedriver

# Open CSV file to write data
with open('./data/nhatot.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "KM", "Price"])

    # Loop through pages
    for i in range(1, 51):
        url = f"https://www.nhatot.com/mua-ban-dat-da-nang?page={i}"
        driver.get(url)

        # Parse page source with BeautifulSoup
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Find all ads
        ads = soup.find_all('div', class_='AdBody_adUpperBody__4Mt0b')

        for ad in ads:
            title = ad.find('h3', class_='commonStyle_adTitle__g520j').text.strip()
            km = ad.find('div', class_='AdBody_adUpperBody__4Mt0b').text.strip()
            price = ad.find('div', class_='AdBody_adItemPrice__Xr5sP').text.strip()

            # Write data to CSV
            writer.writerow([title, km, price])

# Close the driver
driver.quit()
```
